## 정확성

```python
df = pd.read_csv('data/beer.csv', index_col = 0)
df
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
df['abv'].quantile(0.75), df['abv'].quantile(0.5), df['abv'].quantile(0.25)
```

```python
# 이상치 탐색

q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
# 이상치 제거
df.drop(df[condition].index, inplace=True)
```

```python
df.plot(kind='box', y='abv')
```
