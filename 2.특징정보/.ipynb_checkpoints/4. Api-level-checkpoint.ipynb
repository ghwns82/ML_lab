# 개요

* 안드로이드는 버전을 업그레이드 하면서 permission이나 api 등을 추가하거나 삭제한다.
* 안드로이드의 버전은 플랫폼버전, api level, version_code가 있음
* 안드로이드는 버전 코드를 컵케익, 도넛, 젤리, 사탕 같이 귀엽게 지음 ^^
* api level은 작성기준(23.01.12) 33까지 나옴



* api level 종류
    * api level은 AndroidManifest.xml에 기술되어 있다.
    * minSdkVersion : 애플리케이션이 실행하는 데 필요한 최소 API 수준
    * targetSdkVersion : 애플리케이션의 대상 API 수준을 지정하는 정수입니다. 설정하지 않을 경우 기본값은 minSdkVersion에 주어진 값과 동일
    * maxSdkVersion : 애플리케이션을 실행하도록 설계된 최대 API 수준을 지정(그런데 연구엔 딱히 필요없음)
    * complileSdkVersion : 컴파일 시 사용되는 Android API 버전. 실제 개발중 사용할 수 있는 android API 범위는 compileSdkVersion 에 의해 결정됨. compileSdkVersion 값은 가급적 최신으로 유지하기를 권장하며, targetSdkVersion보다 낮을 경우 오류가 나며 실행이 되지않음
    * platformBuildVersionCode : 컴파일러에 의해 추가되며, 앱이 빌드되는 프레임워크 리소스의 버전코드임.
        * 참조 : https://android.googlesource.com/platform/frameworks/base/+/ad2d07d
