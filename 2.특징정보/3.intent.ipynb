# 개요

* intent란 component를 실행하기 위해 시스템에 넘기는 정보
* 컴포넌트간 통신을 위한 메신저로 사용됨
* 인텐트는 명시적 인텐트(Explicit Intent)와 암시적 인텐트(Implicit Intent)로 구분됨
    * 명시적 인텐트 : 인텐트에 클래스 객체나 컴포넌트 이름을 지정하여 호출될 대상을 확실히 알 수 있는 인텐트
    * 암시적 인텐트 : 호출될 대상의 속성들을 지정했지만 호출될 대상이 달라질 수 있는 인텐트

# 추출

## androguard 사용

```python
from androguard.misc import AnalyzeAPK
a, *_ = AnalyzeAPK('/home/ghwns82/ap.apk')

i_list =[]
for act in a.get_activities():
    filters = a.get_intent_filters("activity", act)
    if filters:
        i_list.append((act, filters))
for rec in a.get_receivers():
    filters = a.get_intent_filters("receiver", rec)
    if filters:
        i_list.append((rec, filters))
for ser in a.get_services():
    filters = a.get_intent_filters("service", ser)
    if filters:
        i_list.append((ser, filters))
```
