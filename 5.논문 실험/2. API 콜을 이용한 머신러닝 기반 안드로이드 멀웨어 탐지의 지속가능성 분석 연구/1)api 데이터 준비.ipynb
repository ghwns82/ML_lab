```python
if not os.path.isdir('./csv'):
    os.mkdir('./csv')
    
for year in years:
    for apk_type in apk_types:
        df = pd.read_csv(f'/mnt/z/0.Dataset/AndroZoo/{year}/features/api/{apk_type[:3]}_api_{year}.csv')
        print(year, apk_type, df.shape)
        df.to_csv(f'./csv/{apk_type}_{year}.csv')
```

```python
if not os.path.isdir('./csv'):
    os.mkdir('./csv')
for year in years:
    df_ben = pd.read_csv(f'/mnt/z/0.Dataset/AndroZoo/{year}/features/api/ben_api_{year}.csv', index_col=0)
    df_mal = pd.read_csv(f'/mnt/z/0.Dataset/AndroZoo/{year}/features/api/mal_api_{year}.csv', index_col=0)
    min_ = min(df_ben.shape[0], df_mal.shape[0])
    df_ben = df_ben.sample(n=min_)
    df_mal = df_mal.sample(n=min_)
    
    df_ben.to_csv(f'./csv/{apk_types[0]}_{year}.csv')
    df_mal.to_csv(f'./csv/{apk_types[1]}_{year}.csv')
    
    print(year, 'done')
```
