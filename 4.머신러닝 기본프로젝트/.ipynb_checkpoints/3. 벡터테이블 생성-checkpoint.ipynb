# 개요

* 추출한 특징정보를 기준에 따라 벡터테이블로 변환하는 작업
* 벡터테이블에 permission은 존재여부, api는 빈도를 작성한다
* 데이터에 노이즈가 있는 경우 노이즈 제거가 필요하다




# 코드

```python
import os, csv
```

```python
if not os.path.isdir('./benign/csv'):
    os.mkdir('./benign/csv')
if not os.path.isdir('./malware/csv'):
    os.mkdir('./malware/csv')
```

```python
with open('./data/1848_api.txt') as f:
    api_1848 = f.read().split('\n')
with open('./data/474_perm.txt') as f:
    perm_474 = f.read().split('\n')
```

```python
perm_dict=dict((p,i) for i, p in enumerate(perm_474))
api_dict=dict((a,i) for i, a in enumerate(api_1848))
```

permission vector table

```python
perm_vector_table = []
for file in os.listdir('./benign/permission/'):
    perm_list = [0]*474
    with open(f'./benign/permission/{file}', 'r') as f:
        apk=f.read().split('\n')[1:-1]
    
    for line in apk:
        # 노이즈 제거
        if "'" in line:
            perm = line.split("'")[1]
        elif 'name=' in line:
            perm = line.split("name=")[1]
        else:
            perm = line.split()[1]
        # 벡터테이블 생성
        if perm in perm_474:
            perm_list[perm_dict[perm]]=1
    perm_vector_table.append(perm_list)
    
with open("./benign/csv/benign_perm.csv", 'w', newline='') as f:
    write = csv.writer(f)
    write.writerows(perm_vector_table)
```

```python
perm_vector_table = []
for file in os.listdir('./malware/permission/'):
    perm_list = [0]*474
    with open(f'./malware/permission/{file}', 'r') as f:
        apk=f.read().split('\n')[1:-1]
    for line in apk:
        if "'" in line:
            perm = line.split("'")[1]
        elif 'name=' in line:
            perm = line.split("name=")[1]
        else:
            perm = line.split()[1]
    for perm in apk:
        if perm in perm_474:
            perm_list[perm_dict[perm]]=1
    perm_vector_table.append(perm_list)
    
with open("./malware/csv/malware_perm.csv", 'w', newline='') as f:
    write = csv.writer(f)
    write.writerows(perm_vector_table)
```
