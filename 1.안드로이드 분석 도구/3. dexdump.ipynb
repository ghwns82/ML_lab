# 개요

* dexdump란 안드로이드의 독자적인 VM인 Dalvik을 분석하는 툴임.
* .dex 파일을 조작하여 파일 헤더의 정보를 취득하거나 바이너리 코드를 역어셈블하는 등 다양한 기능이 존재한다
* 안드로이드 sdk 버전에 따라 sdk내부에 dexdump나 dexdump2가 있음

# 설치
