# Apktool

## 개요

* 대표적인 안드로이드 애플리케이션(Android Application) 대상의 리버스 엔지니어링 도구
* 디코딩 : apk 파일을 분석하여 리소스를 뽑을 수 있음
* 리빌드 : 코드를 수정하여 다시 재빌드함

## 설치

https://ibotpeaches.github.io/Apktool/install/ 에 잘 설명되어있음 (23.01.06)

### window
1. 최신 apktool.jar과 apktool.bat을 다운로드한다.

2. apktool.bat의 내용은 아래 링크에 있음. 메모장에 copy & paste 하여 apktool.bat으로 저장후 사용하면 됨. 링크가 유효하지 않을 시 최신버전을 검색하기 바람

https://raw.githubusercontent.com/iBotPeaches/Apktool/master/scripts/windows/apktool.bat

3. 다음 명령어로 실행 (경로 귀찮으면 환경변수 설정하면 된다)

`java -jar {apktool.jar 파일이름}`


## Linux(ubuntu)
1. 완전 쉽다
2. `sudo apt-get install apktool`하면 완료됨
3. `apktool` 명령어로 실행가능, 윈도우와 달리 경로문제도 없음


## 사용법

크게 decompile(디코딩)과 build(리빌드)가 있음

### 디코딩
`apktool d -o 저장경로 apk파일`

ap.apk 파일을 디코딩한다. 디코딩한 결과는 decoded_myap로 저장됨


### 리빌드
`apktool b -o myap.apk myap`

decoded_ap 디렉토리 내용을 converted_ap.apk의 APK 파일로 리빌드



# aapt

## 개요

* Android Asset Packaging Tool
* Android 스튜디오 및 Android Gradle 플러그인이 앱의 리소스를 컴파일하고 패키징하는 데 사용하는 빌드 도구
* aapt는 다양한 버전이 존재하며 aapt2도 따로 존재함. 작성날짜기준(23.01.06) aapt2를 많이 쓰는 것으로 생각됨
* aapt와 aapt2는 비슷하나, 명령어나 출력결과에 일부 차이가 있음

## 설치

### Android Studio에 내장된 aapt 가져오기
* android studio 설치
* android studio 디렉토리에서 aapt 탐색
* 없는 경우도 있음

### apt
* `sudo apt install aapt` 실행 (linux)
* `pip install aapt` 실행 (Window)

### android sdk에서 추출
* 가장 확실한 방법
* 다소 어려움
* 가장 최신버전 추출 가능

## 사용방법(api level, permission 추출)

window는 aapt.exe, Linux는 aapt로 실행

aapt는 dump, pakage 등 다양한 기능이 있으나, 보통 dump만 쓰며 그 중 permissions와 badging만 주로 사용


그 외 옵션은 `aapt`나 `aapt2`를 입력하면 description이 있으니 필요하다면 여러가지 시도해보면 좋다




# dexdump

## 개요

* dexdump란 안드로이드의 독자적인 VM인 Dalvik을 분석하는 툴임.
* .dex 파일을 조작하여 파일 헤더의 정보를 취득하거나 바이너리 코드를 역어셈블하는 등 다양한 기능이 존재한다
* 안드로이드 sdk 버전에 따라 sdk내부에 dexdump나 dexdump2가 있음

## 설치

### apt
* `sudo apt install dexdump` 실행 (linux)
